# Binning

```python
data['SEC_CONTRIBUTORY_CAUSE_N'] = data['SEC_CONTRIBUTORY_CAUSE'].apply(
    lambda x: 'UNKNOWN' if x in ['UNABLE TO DETERMINE', 'NOT APPLICABLE'] 
    else 'WEATHER' if x in [
        'WEATHER']
    
    else 'RECKLESS_DRIVING' if x in [
        'OPERATING VEHICLE IN ERRATIC, RECKLESS, CARELESS, NEGLIGENT OR AGGRESSIVE MANNER', 'IMPROPER OVERTAKING/PASSING', 'IMPROPER BACKING', 'IMPROPER LANE USAGE', 'IMPROPER TURNING/NO SIGNAL', 'FOLLOWING TOO CLOSELY', 'FAILING TO YIELD RIGHT-OF-WAY', 'DRIVING ON WRONG SIDE/WRONG WAY', 'EXCEEDING AUTHORIZED SPEED LIMIT']
    
    else 'DISREGARDING_ROADSIGNS' if x in [
        'EXCEEDING SAFE SPEED FOR CONDITIONS', 'DISREGARDING STOP SIGN', 'DISREGARDING OTHER TRAFFIC SIGNS', 'DISREGARDING ROAD MARKINGS', 'DISREGARDING YIELD SIGN', 'DISREGARDING TRAFFIC SIGNALS', 'PASSING STOPPED SCHOOL BUS', 'RELATED TO BUS STOP']
    
    else 'DISTRACTED' if x in [
        'DISTRACTION - FROM INSIDE VEHICLE', 'DISTRACTION - FROM OUTSIDE VEHICLE', 'CELL PHONE USE OTHER THAN TEXTING', 'TEXTING', 'DISTRACTION - OTHER ELECTRONIC DEVICE (NAVIGATION DEVICE, DVD PLAYER, ETC.)']
    
    else 'UNDER_INFLUENCE' if x in [
        'UNDER THE INFLUENCE OF ALCOHOL/DRUGS (USE WHEN ARREST IS EFFECTED)', 'HAD BEEN DRINKING (USE WHEN ARREST IS NOT MADE)', 'PHYSICAL CONDITION OF DRIVER']
    
    else 'ROAD_ENGINEERING' if x in [
        'ROAD ENGINEERING/SURFACE/MARKING DEFECTS', 'ROAD CONSTRUCTION/MAINTENANCE', 'VISION OBSCURED (SIGNS, TREE LIMBS, BUILDINGS, ETC.)', 'OBSTRUCTED CROSSWALKS', 'ANIMAL']
    
    else 'DRIVING_EXPERIENCE' if x in ['BICYCLE ADVANCING LEGALLY ON RED LIGHT', 'MOTORCYCLE ADVANCING LEGALLY ON RED LIGHT', 'TURNING RIGHT ON RED', 'DRIVING SKILLS/KNOWLEDGE/EXPERIENCE', 'FAILING TO REDUCE SPEED TO AVOID CRASH', 'EQUIPMENT - VEHICLE CONDITION', 'EVASIVE ACTION DUE TO ANIMAL, OBJECT, NONMOTORIST']
    
    else 'OTHER'
    )
```

```python
data['SEC_CONTRIBUTORY_CAUSE_N'].value_counts()
```

```python
data['PRIM_CONTRIBUTORY_CAUSE_N'] = data['PRIM_CONTRIBUTORY_CAUSE'].apply(
    lambda x: 'UNKNOWN' if x in ['UNABLE TO DETERMINE', 'NOT APPLICABLE'] 
    
    else 'WEATHER' if x in [
        'WEATHER']
    
    else 'RECKLESS_DRIVING' if x in [
        'OPERATING VEHICLE IN ERRATIC, RECKLESS, CARELESS, NEGLIGENT OR AGGRESSIVE MANNER', 'IMPROPER OVERTAKING/PASSING', 'IMPROPER BACKING', 'IMPROPER LANE USAGE', 'IMPROPER TURNING/NO SIGNAL', 'FOLLOWING TOO CLOSELY', 'FAILING TO YIELD RIGHT-OF-WAY', 'DRIVING ON WRONG SIDE/WRONG WAY', 'EXCEEDING AUTHORIZED SPEED LIMIT']
    
    else 'DISREGARDING_ROADSIGNS' if x in [
        'EXCEEDING SAFE SPEED FOR CONDITIONS', 'DISREGARDING STOP SIGN', 'DISREGARDING OTHER TRAFFIC SIGNS', 'DISREGARDING ROAD MARKINGS', 'DISREGARDING YIELD SIGN', 'DISREGARDING TRAFFIC SIGNALS', 'PASSING STOPPED SCHOOL BUS', 'RELATED TO BUS STOP']
    
    else 'DISTRACTED' if x in [
        'DISTRACTION - FROM INSIDE VEHICLE', 'DISTRACTION - FROM OUTSIDE VEHICLE', 'CELL PHONE USE OTHER THAN TEXTING', 'TEXTING', 'DISTRACTION - OTHER ELECTRONIC DEVICE (NAVIGATION DEVICE, DVD PLAYER, ETC.)']
    
    else 'UNDER_INFLUENCE' if x in [
        'UNDER THE INFLUENCE OF ALCOHOL/DRUGS (USE WHEN ARREST IS EFFECTED)', 'HAD BEEN DRINKING (USE WHEN ARREST IS NOT MADE)', 'PHYSICAL CONDITION OF DRIVER']
    
    else 'ROAD_ENGINEERING' if x in [
        'ROAD ENGINEERING/SURFACE/MARKING DEFECTS', 'ROAD CONSTRUCTION/MAINTENANCE', 'VISION OBSCURED (SIGNS, TREE LIMBS, BUILDINGS, ETC.)', 'OBSTRUCTED CROSSWALKS', 'ANIMAL']
    
    else 'DRIVING_EXPERIENCE' if x in ['BICYCLE ADVANCING LEGALLY ON RED LIGHT', 'MOTORCYCLE ADVANCING LEGALLY ON RED LIGHT', 'TURNING RIGHT ON RED', 'DRIVING SKILLS/KNOWLEDGE/EXPERIENCE', 'FAILING TO REDUCE SPEED TO AVOID CRASH', 'EQUIPMENT - VEHICLE CONDITION', 'EVASIVE ACTION DUE TO ANIMAL, OBJECT, NONMOTORIST']
    
    else 'OTHER'
)
```

```python
data['PRIM_CONTRIBUTORY_CAUSE_N'].value_counts()
```

```python
data['TRAFFICWAY_TYPE_N'] = data['TRAFFICWAY_TYPE'].apply(
    lambda x: 'NOT_DIVIDED' if x in [
        'NOT DIVIDED'
    ]
    
    else 'DIVIDED' if x in [
        'DIVIDED - W/MEDIAN (NOT RAISED)', 'DIVIDED - W/MEDIAN BARRIER'
    ]
    
    else 'ONE_WAY' if x in [
        'ONE-WAY'
    ]
    
    else 'PARKED' if x in [
        'PARKING LOT', 'DRIVEWAY'
    ]
    
    else 'UNKNOWN' if x in [
        'OTHER', 'UNKNOWN', 'TRAFFIC ROUTE'
    ]
    
    else 'ALLEY' if x in [
        'ALLEY'
    ]
    
    else 'INTERSECTION' if x in [
        'T-INTERSECTION', 'Y-INTERSECTION', 'L-INTERSECTION', 'UNKNOWN INTERSECTION TYPE', 'FOUR WAY', 'RAMP'
    ]
    
    else 'CENTER_TURN_LANE' if x in [
        'CENTER TURN LANE'
    ]
    
    else 'ROUNDABOUT' if x in [
        'FIVE POINT, OR MORE', 'ROUNDABOUT'
    ]
    
    else 'NOT_REPORTED' if x in [
        'NOT REPORTED'
    ]
    
    else 'OTHER'
)
```

```python
data['TRAFFICWAY_TYPE_N'].value_counts()
```

```python
# road surface

data['ROAD_DEFECT_N'] = data['ROAD_DEFECT'].apply(
    lambda x: 'NO_DEFECTS' if x in [
        'NO DEFECTS'
    ]
    
    else 'UNKNOWN' if x in [
        'UNKNOWN', 'OTHER'
    ]
    
    else 'WORN_SURFACE' if x in [
        'RUT, HOLES', 'WORN SURFACE'
    ]
    
    else 'SHOULDER_DEFECT' if x in [
        'SHOULDER DEFECT'
    ]
    
    else 'DEBRIS' if x in [
        'DEBRIS ON ROADWAY'
    ]
    
    else 'OTHER'
)
```

```python
data['ROAD_DEFECT_N'].value_counts()
```

```python
# road surface
data['ROADWAY_SURFACE_COND_N'] = data['ROADWAY_SURFACE_COND'].apply(
    lambda x: 'DRY' if x in [
        'DRY'
    ]
    
    else 'WET' if x in [
        'WET'
    ]
    
    else 'UNKNOWN' if x in [
        'UNKNOWN', 'OTHER'
    ]
    
    else 'SNOW' if x in [
        'SNOW OR SLUSH'
    ]
    
    else 'ICE' if x in [
        'ICE'
    ]
    
    else 'SAND' if x in [
        'SAND, MUD, DIRT'
    ]
    
    else 'OTHER'
)
```

```python
data['ROADWAY_SURFACE_COND_N'].value_counts()
```

```python
# first crash
data['FIRST_CRASH_TYPE_N'] = data['FIRST_CRASH_TYPE'].apply(
    lambda x: 'PARKED' if x in [
        'PARKED MOTOR VEHICLE'
    ]
    
    else 'REAR_END' if x in [
        'REAR END', 'REAR TO REAR', 'REAR TO FRONT'
    ]
    
    else 'SIDESWIPE' if x in [
        'SIDESWIPE SAME DIRECTION' , 'SIDESWIPE OPPOSITE DIRECTION'
    ]
    
    else 'ANGLE' if x in [
        'TURNING', 'ANGLE', 'REAR TO SIDE'
    ]
    
    else 'OBJECT' if x in [
        'FIXED OBJECT', 'OTHER OBJECT'
    ]
    
    else 'PEDESTRIAN' if x in [
        'PEDESTRIAN'
    ]
    
    else 'PEDALCYCLIST' if x in [
        'PEDALCYCLIST'
    ]
    
    else 'HEAD_ON' if x in [
        'HEAD ON'
    ]
    
    else 'NON_COLLISION' if x in [
        'OTHER NONCOLLISION', 'OVERTURNED'
    ]
    
    else 'ANIMAL' if x in [
        'ANIMAL'
    ]
    
    else 'TRAIN' if x in [
        'TRAIN'
    ]
    
    else 'OTHER'
)
```

```python
data['FIRST_CRASH_TYPE_N'].value_counts()
```

```python
data['WEATHER_CONDITION_N'] = data['WEATHER_CONDITION'].apply(
    lambda x: 'CLEAR' if x in [
        'CLEAR'
    ]
    
    else 'RAIN' if x in [
        'RAIN', 'FREEZING RAIN/DRIZZLE'
    ]
    
    else 'UNKNOWN' if x in [
        'UNKNOWN', 'OTHER'
    ]
    
    else 'SNOW' if x in [
        'SNOW'
    ]
    
    else 'CLOUD' if x in [
        'CLOUDY/OVERCAST'
    ]
    
    else 'LOW_VISIBILITY' if x in [
        'FOG/SMOKE/HAZE', 'BLOWING SAND, SOIL, DIRT', 'BLOWING SNOW'
    ]
    
    else 'SLEET' if x in [
        'SLEET/HAIL'
    ]
    
    else 'STRONG_WIND' if x in [
        'SEVERE CROSS WIND GATE'
    ]
    
    else 'OTHER'
)
```
