# EDA

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load in

```python
data = pd.read_csv('CLEANDATA/clean_data_N.csv', index_col=False)
data.head().T
```

## drop columns

```python
# WEATHER_CONDITION, FIRST_CRASH_TYPE, TRAFFICWAY_TYPE, ROADWAY_SURFACE_COND, ROAD_DEFECT, PRIM_CONTRIBUTORY_CAUSE, SEC_CONTRIBUTORY_CAUSE, TRAFFIC_CONTROL_DEVICE, DEVICE_CONDITION

data = data.drop(columns=['WEATHER_CONDITION', 'FIRST_CRASH_TYPE', 'TRAFFICWAY_TYPE', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'PRIM_CONTRIBUTORY_CAUSE', 'SEC_CONTRIBUTORY_CAUSE', 'TRAFFIC_CONTROL_DEVICE', 'DEVICE_CONDITION'])
data.info()
```

```python
# STREET_NO, STREET_DIRECTION, STREET_NAME, BEAT_OF_OCCURRENCE

data = data.drop(columns=['STREET_NO', 'STREET_DIRECTION', 'STREET_NAME', 'BEAT_OF_OCCURRENCE'])
data.info()
```

```python
# INJURIES_TOTAL, INJURIES_FATAL, INJURIES_INCAPACITATING, INJURIES_NON_INCAPACITATING, INJURIES_REPORTED_NOT_EVIDENT, INJURIES_NO_INDICATION, INJURIES_UNKNOWN

data = data.drop(columns=['INJURIES_TOTAL', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_NON_INCAPACITATING', 'INJURIES_REPORTED_NOT_EVIDENT', 'INJURIES_NO_INDICATION', 'INJURIES_UNKNOWN'])
data.info()
```

```python
data.head().T
```

## Visualize

```python
sns.violinplot(y='LIGHTING_CONDITION',x='DAMAGE', data=data)
plt.show()
```

```python
# Set up the figure
plt.figure(figsize=(20, 15))

# 1. Count Plot for Lighting Condition vs. Damage
plt.subplot(3, 2, 1)
sns.countplot(data=data, x='LIGHTING_CONDITION', hue='DAMAGE')
plt.title('Count of Damage by Lighting Condition')
plt.xticks(rotation=45)

# 2. Count Plot for Crash Type vs. Damage
plt.subplot(3, 2, 2)
sns.countplot(data=data, x='CRASH_TYPE', hue='DAMAGE')
plt.title('Count of Damage by Crash Type')
plt.xticks(rotation=45)

# 3. Count Plot for Most Severe Injury vs. Damage
plt.subplot(3, 2, 3)
sns.countplot(data=data, x='MOST_SEVERE_INJURY', hue='DAMAGE')
plt.title('Count of Damage by Most Severe Injury')
plt.xticks(rotation=45)

# 4. Count Plot for Trafficway Type vs. Damage
plt.subplot(3, 2, 4)
sns.countplot(data=data, x='TRAFFICWAY_TYPE_N', hue='DAMAGE')
plt.title('Count of Damage by Trafficway Type')
plt.xticks(rotation=45)

# 5. Count Plot for Weather Condition vs. Damage
plt.subplot(3, 2, 5)
sns.countplot(data=data, x='WEATHER_CONDITION_N', hue='DAMAGE')
plt.title('Count of Damage by Weather Condition')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
